```python
import pandas as pd
import matplotlib.pyplot as plt

everland = pd.read_csv('EVERLAND(2006~2022)입장객.csv', encoding='cp949', thousands=',')
everland.set_index('월별')['2022년'][:12].plot(kind='bar')

plt.title('2022년 에버랜드 방문객')
plt.rc('font', family='Malgun Gothic')
plt.style.use('fivethirtyeight')
plt.xlabel('')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

everland = pd.read_csv('EVERLAND(2006~2022)입장객.csv', encoding='cp949', thousands=',')

everland.set_index('월별')['2022년'][:12].plot(label='2022년(점차 회복)')
everland.set_index('월별')['2020년'][:12].plot(label='2020년(코로나 유행)', color='r', marker='o')

plt.title('코로나에 따른 방문객 감소')
plt.rc('font', family='Malgun Gothic')
plt.style.use('fivethirtyeight')
plt.xlabel('')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

everland = pd.read_csv('EVERLAND(2006~2022)입장객.csv', encoding='cp949', thousands=',')

everland.set_index('월별')[['2017년','2018년','2019년']][:12].plot()

plt.title('EVERLAND(2017년~2019년)')
plt.rc('font', family='Malgun Gothic')
plt.style.use('fivethirtyeight')
plt.xlabel('')
plt.legend()
plt.show()
```
