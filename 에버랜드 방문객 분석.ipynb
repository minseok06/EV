```python
import pandas as pd
import matplotlib.pyplot as plt

everland = pd.read_csv('EVERLAND(2006~2022)입장객.csv', encoding='cp949', thousands=',')
everland.set_index('월별')['2022년'][:12].plot(kind='bar')

plt.title('2022년 에버랜드 방문객')
plt.rc('font', family='Malgun Gothic')
plt.style.use('fivethirtyeight')
plt.xlabel('')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

everland = pd.read_csv('EVERLAND(2006~2022)입장객.csv', encoding='cp949', thousands=',')

everland.set_index('월별')['2022년'][:12].plot(label='2022년(점차 회복)')
everland.set_index('월별')['2020년'][:12].plot(label='2020년(코로나 유행)', color='r', marker='o')

plt.title('코로나에 따른 방문객 감소')
plt.rc('font', family='Malgun Gothic')
plt.style.use('fivethirtyeight')
plt.xlabel('')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

everland = pd.read_csv('EVERLAND(2006~2022)입장객.csv', encoding='cp949', thousands=',')

ev_2017 = everland.set_index('월별')['2017년'][:12]
ev_2018 = everland.set_index('월별')['2018년'][:12]
ev_2019 = everland.set_index('월별')['2019년'][:12]
ev_2020 = everland.set_index('월별')['2020년'][:12]
ev_2021 = everland.set_index('월별')['2021년'][:12]
ev_2022 = everland.set_index('월별')['2022년'][:12]


plt.figure(figsize=(12,5))
plt.plot(ev_2017, label='2017')
plt.plot(ev_2018, label='2018')
plt.plot(ev_2019, label='2019')
plt.plot(ev_2020, label='2020(코로나)')
plt.plot(ev_2021, label='2021')
plt.plot(ev_2022, label='2022')
plt.title('EVERLAND(2017년~2022년) 방문객')
plt.rc('font', family='Malgun Gothic')
plt.style.use('fivethirtyeight')
plt.legend()
plt.show()
```
